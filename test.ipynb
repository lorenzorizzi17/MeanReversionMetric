```python
import nbimporter
import numpy as np
import matplotlib.pyplot as plt
import data_collection as dc

# let's start with four dataset
DAX = dc.load_asset("data/DAX_M15_202005242300_202411131015.csv")
DJ = dc.load_asset("data/DJ_M15_202009240845_202411131015.csv")
EURUSD = dc.load_asset("data/EURUSD_M15_202012241045_202412311745.csv")
FTSE = dc.load_asset("data/FTSE100_M15_201912230600_202411131015.csv")

# if we wanted to plot them, then
# dc.draw_info(DAX)
```

```python
# SPLITTING IN TRENDS
# Ho aggiunto anche i plot solo per avere conferma visiva, chiaramente si possono eliminare, la parte importante in questa cella sono gli arr[...]
fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize=(15,10))

# splitting in trends
meshDAX = np.array([0,int(len(DAX)/4), int(len(DAX)/2), int(len(DAX))])
arrDAX = dc.removeTrend(DAX, meshDAX, ax[0,0])
ax[0,0].plot(DAX.index, DAX["Close"], c = "g")

meshDJ = np.array([0,int(len(DJ)/4), int(len(DJ)/2), int(len(DJ))])
arrDJ = dc.removeTrend(DJ, meshDJ, ax[0,1])
ax[0,1].plot(DJ.index, DJ["Close"], c = "g")

meshEURUSD = np.array([0, int(2.15*len(EURUSD)/5), int(2.75*len(EURUSD)/5), int(len(EURUSD))])
arrEURUSD = dc.removeTrend(EURUSD, meshEURUSD, ax[1,0])
ax[1,0].plot(EURUSD.index, EURUSD["Close"], c = "g")

meshFTSE = np.array([int(0.2*len(FTSE)/4), int(len(FTSE)/4),int(len(FTSE))])
arrFTSE = dc.removeTrend(FTSE, meshFTSE, ax[1,1])
ax[1,1].plot(FTSE.index, FTSE["Close"], c = "g")

plt.show()
```
